```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename_u = "./build/results/u_197811.txt"
filename_p = "./build/results/p_197811.txt"
filename_v = "./build/results/v_197811.txt"

filename_c = "./build/params.txt"

with open (filename_c) as file:
    data = np.fromstring(file.read(), sep=' ')
    L = data[0]
    NX = data[1]
    NY = data[2]
    RE = data[3]
    HT = data[4]
    #print(p)

with open (filename_p) as file:
    data = file.read()
    p = np.fromstring(data, sep=' ')
    #print(p)

with open (filename_u) as file:
    data = file.read()
    u = np.fromstring(data, sep=' ')
    #print(p)

with open (filename_v) as file:
    data = file.read()
    v = np.fromstring(data, sep=' ')
    #print(p)

N = (int) (len(p) ** (1/2))

u = np.flip(u.reshape(N, N))
v = v.reshape(N, N)
p = p.reshape(N, N)


#   
x = np.linspace(0, 1, N)
y = np.linspace(0, 1, N)

X, Y = np.meshgrid(x, y)

title = "           L = " + str(L) + " NX = " + str(NX) + " NY = " + str(NY) + " RE = " + str(RE) + " HT = " + str(HT)
fig, ax = plt.subplots()
ax.streamplot(X, Y, u, v)
im = ax.imshow(p, cmap='plasma', extent=[0, 1, 0, 1])
fig.colorbar(im)
plt.title(label=title)
```
